```python
import geopandas as gpd
import os
import shutil
```

```python
root = r"D:\cassda-data\street-tiles\DE\images"
export_dir = r"D:\cassda-data\street-tiles\DE\filtered-background-imgs"
```

```python
gdf = gpd.read_file("../model/runs/cassda-wildlife-crossing/yolo11-DE-conf0_35/detection_results_filtered.csv")
```

```python
for idx, row in gdf.iterrows():
    if idx > 100:
        break
    img_path = row["image_path"]
    img_path = os.path.join(root, img_path)
    export_path = os.path.join(export_dir, os.path.relpath(img_path, root))
    os.makedirs(os.path.dirname(export_path), exist_ok=True)
    shutil.copy(img_path, export_path)
    print(f"Copied {img_path} to {export_path}")
```

```python
import glob

for img in glob.glob(r"D:\cassda-data\street-tiles\DE\filtered-background-imgs\images\*.jpg"):
    with open(img.replace("images", "labels").replace(".jpg", ".txt"), "w") as f:
        f.write("")
```
